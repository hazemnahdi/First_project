```python
from google.colab import files
from openpyxl import Workbook
import pandas as pd
import ipywidgets as widgets
import io
import os
import numpy as np

# Fonction pour gérer le téléchargement de fichiers
def telecharger_fichiers(change):
    print("Sélectionnez les fichiers à traiter.")
    fichiers = files.upload()
    dossier_resultats = 'resultats_filtres'
    if not os.path.exists(dossier_resultats):
        os.makedirs(dossier_resultats)

    # Définition des valeurs à filtrer pour les lignes spécifiées
    valeurs_a_filtrer = {
        'TIRAGE': ('21', '281', '231', '291'),
        'Bilan': (' Immobilisations incorporelles',)
    }

    for nom_fichier, contenu_fichier in fichiers.items():
        print(f"Fichier téléchargé : {nom_fichier}")

        # Charger le fichier Excel dans un DataFrame
        dataframe_etats_financiers = pd.read_excel(io.BytesIO(contenu_fichier))

        # Nettoyer les noms de colonnes
        dataframe_etats_financiers.columns = dataframe_etats_financiers.columns.str.strip()

        # Filtrer les lignes selon les valeurs spécifiées en fonction du nom du fichier
        if any(nom_fichier.startswith(nom) for nom in valeurs_a_filtrer.keys()):
            valeurs_a_filtrer_nom = valeurs_a_filtrer[nom_fichier.split('_')[0]]
            colonne_1_filtrée = dataframe_etats_financiers[dataframe_etats_financiers.iloc[:, 0].astype(str).apply(lambda x, valeurs: any(x.startswith(val) for val in valeurs), valeurs=valeurs_a_filtrer_nom)]

            # Nom du fichier résultat
            nom_fichier_resultat = os.path.join(dossier_resultats, f'rapport_audit_{nom_fichier}')

            # Charger le classeur Excel existant ou créer un nouveau classeur
            try:
                workbook = Workbook()
            except FileNotFoundError:
                workbook = Workbook()

            # Créer une nouvelle feuille
            feuille = workbook.active
            feuille.title = 'Résultats'

            # Ajouter les données filtrées à la feuille
            for row in colonne_1_filtrée.values:
                feuille.append(list(row))  # Convertir le ndarray en liste

            # Sauvegarder le classeur Excel
            workbook.save(nom_fichier_resultat)

            print(f"Résultats ajoutés à '{nom_fichier_resultat}'")

    # Zipper les fichiers résultants dans un seul fichier
    nom_zip = os.path.join(dossier_resultats, 'resultats.zip')
    os.system(f'zip -r {nom_zip} {dossier_resultats}')

    # Télécharger le fichier zip contenant les résultats
    files.download(nom_zip)

# Ajouter un bouton pour importer les fichiers
bouton_importation = widgets.Button(description="Importer les fichiers")
bouton_importation.on_click(telecharger_fichiers)

# Afficher le bouton
display(bouton_importation)
```
